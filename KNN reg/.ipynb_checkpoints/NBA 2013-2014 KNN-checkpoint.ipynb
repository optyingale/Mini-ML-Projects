# Problem Statement:
**Predict how many points a NBA player would score in 2013-2014 season**

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split, GridSearchCV

import seaborn as sns
import matplotlib.pyplot as plt

# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('nba_2013.csv')
```

```python
# Filling null values by 0 because thats what it actually means for this data

df.fillna(0, inplace=True)
na = dict(df.isna().sum())
for x in na:
    if na[x] != 0:
        print(x, na[x])

df.drop(['bref_team_id', 'season', 'season_end'], axis=1, inplace=True)
```

```python
print(df['pos'].value_counts(), end='\n\n')
print(df['pos'].unique(), '\nnumber of unique values = ', len(df['pos'].unique()))
```

```python
onh = preprocessing.OneHotEncoder()
X = df['pos']
X = onh.fit_transform(X.values.reshape(-1,1)).toarray()
onhdf = pd.DataFrame(X.astype(int), columns = ["pos_"+str(int(i)) for i in range(len(df['pos'].unique()))])
df = pd.concat([df, onhdf], axis=1)
for i in onh.categories_:
    print(i)
```

```python
df.drop(['player', 'pos'], axis=1, inplace=True)
```

### Checking Distribution of continuous data

```python
columns = list(df.describe().columns)

plt.figure(figsize=(20,50),facecolor='white')
plotnumber = 1

for column in columns:
    ax = plt.subplot(11, 3, plotnumber)
    sns.distplot(df[column])
    plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

# Decided not to scale the data based on above distribution

```python
X = df.drop('pts', axis=1)
y = df['pts']

k_model = KNeighborsRegressor()
parameters = {'n_neighbors' : range(1,28,2), 
              'weights' : ['uniform', 'distance'], }
             #'leaf_size' : range(1,35)}
#scoring = {'AUC': 'roc_auc', 'Accuracy': make_scorer(accuracy_score)}
scoring = {'recall' : recall_score, 'precision' : precision_score}

clf = GridSearchCV(k_model, parameters, verbose=1, n_jobs=-1, cv=7)
clf.fit(X, y)
cv_result = pd.DataFrame(clf.cv_results_)
print('\n', clf.best_estimator_)
print('\nbest score = ', clf.best_score_)
```
