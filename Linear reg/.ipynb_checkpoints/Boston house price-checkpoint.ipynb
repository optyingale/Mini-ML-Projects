# Problem Statement:
**Linear Regression model to predict house pricing in Boston**

```python
# Importing Necessities

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import ElasticNetCV, LinearRegression
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.datasets import load_boston
import seaborn as sns
from imblearn.over_sampling import RandomOverSampler

# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

## Loading and assigning data

```python
boston = load_boston()
bos_features = pd.DataFrame(boston.data)
bos_target = pd.DataFrame(boston.target, columns=['Target'])
df = pd.concat([bos_features, bos_target], axis=1)
```

## Exploring data

```python
df.info()
```

```python
# No missing data
df.describe()
```

### Distribution of data

```python
columns = list(df.describe().columns)

plt.figure(figsize=(20,20),facecolor='white')
plotnumber = 1

for column in columns:
    ax = plt.subplot(5, 3, plotnumber)
    sns.scatterplot(df[column], df['Target'])
    #sns.distplot(df[column])
    plt.xlabel(column, fontsize=10)
    plotnumber += 1
plt.show()
```

```python

sns.pairplot(df[df.columns], size=3, aspect=1.5, kind='reg',
             #plot_kws=dict(edgecolor="k", linewidth=0.5),
             diag_kind="kde", diag_kws=dict(shade=True));
```

### As we can see "target" is capped at 50, and since this will ruin my prediction model, I'm removing those data points
### Similarly the 11th column is capped at 396.90

```python
df.drop(df[df['Target'] == 50].index.tolist(), inplace=True)
```

```python
df.drop(df[df[11] == 396.90].index.tolist(), inplace=True)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(), annot=True, fmt='.1f')
```

### Removing 3rd column because it has no correlation with any of the columns

```python
df.drop(3, axis=1, inplace=True)
```

## Elastic Net model

```python
# Hold out set for 20% of the data = 75 
hold = np.random.random_integers(0, 368, 75).tolist()
holdout = df.loc[df.index[hold]]
df.drop(df.index[hold], inplace=True)
```

```python
reg = ElasticNetCV(n_jobs=-1, max_iter=10000)
X = df.drop('Target', axis=1)
y = df['Target']
#X, y = ros.fit_resample(X, y)
scoring = {"R**2" : make_scorer(r2_score), 
           "MAE" : make_scorer(mean_absolute_error), 
           "MSE" : make_scorer(mean_squared_error),}
params = {'l1_ratio' : [i/1000 for i in range(0,1000)] }
clf = GridSearchCV(reg, param_grid=params, scoring=scoring, n_jobs=-1, cv=4, verbose=1, refit="R**2")
clf.fit(X,y);
# To know how each of the training examples performed against each test examples
# clf.cv_results_
```

```python
clf.score(holdout.drop('Target', axis=1), holdout['Target'])
```
